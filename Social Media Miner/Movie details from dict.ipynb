```python
import pandas as pd
```

```python
df=pd.read_csv("Cast_Crew_Data.csv")
```

```python
df=df.iloc[:,1:12]
```

```python
df.shape
```

```python
df['All Cast Crew']=df['All Cast Crew'].str.strip()
df['All Cast Crew']=df['All Cast Crew'].str.lstrip()
df['All Cast Crew']=df['All Cast Crew'].str.rstrip()
```

```python
df
```

```python
df.shape
```

```python
df=df[~df['All Cast Crew'].isnull()]
```

```python
df.shape
```

```python
df=df[df['All Cast Crew'].str.contains('{')]
```

```python
df.shape
```

```python
df=df[df['All Cast Crew'].str.contains('}')]
```

```python
df.shape
```

```python
import ast
```

```python
df['cast_dict'] = df['All Cast Crew'].map(ast.literal_eval)
```

```python
import numpy as np
```

```python
df['director_name']=df['cast_dict'].apply(lambda x: x['director'] if 'director' in x else np.nan)
df['writer_name']=df['cast_dict'].apply(lambda x: x['writer'] if 'writer' in x else np.nan)
df['cast_name']=df['cast_dict'].apply(lambda x: x['cast'] if 'cast' in x else np.nan)
df['producer_name']=df['cast_dict'].apply(lambda x: x['producer'] if 'producer' in x else np.nan)
df['composer_name']=df['cast_dict'].apply(lambda x: x['composer'] if 'composer' in x else np.nan)
df['cinematographer_name']=df['cast_dict'].apply(lambda x: x['cinematographer'] if 'cinematographer' in x else np.nan)
df['editor_name']=df['cast_dict'].apply(lambda x: x['editor'] if 'editor' in x else np.nan)
df['casting_director_name']=df['cast_dict'].apply(lambda x: x['casting_director'] if 'casting_director' in x else np.nan)
df['production_designer_name']=df['cast_dict'].apply(lambda x: x['production_designer'] if 'production_designer' in x else np.nan)
df['make_up_department_name']=df['cast_dict'].apply(lambda x: x['make_up_department'] if 'make_up_department' in x else np.nan)
df['production_manager_name']=df['cast_dict'].apply(lambda x: x['production_manager'] if 'production_manager' in x else np.nan)
df['assistant_director_name']=df['cast_dict'].apply(lambda x: x['assistant_director'] if 'assistant_director' in x else np.nan)
df['art_department']=df['cast_dict'].apply(lambda x: x['art_department'] if 'art_department' in x else np.nan)
df['sound_department']=df['cast_dict'].apply(lambda x: x['sound_department'] if 'sound_department' in x else np.nan)
df['visual_effects']=df['cast_dict'].apply(lambda x: x['visual_effects'] if 'visual_effects' in x else np.nan)
df['stunts']=df['cast_dict'].apply(lambda x: x['stunts'] if 'stunts' in x else np.nan)
df['camera_department']=df['cast_dict'].apply(lambda x: x['camera_department'] if 'camera_department' in x else np.nan)
df['music_department']=df['cast_dict'].apply(lambda x: x['music_department'] if 'music_department' in x else np.nan)
df['miscellaneous']=df['cast_dict'].apply(lambda x: x['miscellaneous'] if 'miscellaneous' in x else np.nan)
```

```python
df_movie_release_date=pd.read_excel('movie_list.xlsx')
df_movie_release_date.rename(columns={'Movie_name':'default_movie_name'},inplace=True)
```

```python
df_movie_release_date
```

```python
df_movie_release_date['default_movie_name']=df_movie_release_date['default_movie_name'].astype('str')
df['default_movie_name']=df['default_movie_name'].astype('str')
```

```python
df_movie_release_date['default_movie_name']=df_movie_release_date['default_movie_name'].str.lower()
df['default_movie_name']=df['default_movie_name'].str.lower()
```
